```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Airbnb_data.csv")
df.head()
```

```python
df.columns
```

## Check missing values

```python
print(df.isnull().sum())
```

```python
df.info()
```

## Handling missing values

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month':0,'last review':df['last review'].min()}, inplace=True)
```

```python
df.dropna(subset=["NAME", "host name"],inplace=True)
```

```python
print(df.isnull().sum())
```

## Drop irrelavent columns

```python
df=df.drop(columns=["license", "house_rules"], errors='ignore')
```

```python
df.head()
```

## Removing dollar sign and convert to float datatype

```python
df['price'] = df['price'].replace('[\$,]','', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]',''df, regex=True).astype(float)
```

```python
df.head()
```

## Remove duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

## Descriptive statistics

```python
df.describe()
```

# Visualization

### What is the distribution of listing prices?

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=50, kde= True, color = 'red')
plt.title("Distribution of listing price")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

### How are different room types distributed?

```python
df['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x = "room type", data = df, color= "hotpink")
plt.title("Room type distribution")
plt.xlabel("Room type")
plt.ylabel("Count")
plt.show()
```

### How are listings distributed across different neighbourhood?

```python
plt.figure(figsize=(12,8))
sns.countplot(y="neighbourhood group", data = df, color = 'lightgreen', order=df['neighbourhood group'].value_counts().index)
plt.title("Number of listings by neighbourhood group")
plt.xlabel("Count")
plt.ylabel("Neighbourhood group")
plt.show()
```

### What is the relationship between price and room type?

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="room type", y="price", hue="room type", data = df, palette = 'Set1')
plt.title("Price vs Room type")
plt.xlabel("Room type")
plt.ylabel("Price $")
plt.show()
```

### How has the number of reviews changed over with time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line', color='red')
plt.title("Number of reviews over with time")
plt.xlabel("Date")
plt.ylabel("Number of reviews")
plt.show()
```
